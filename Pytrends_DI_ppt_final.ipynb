```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import calendar
from pytrends.request import TrendReq
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw_list= ['gun']
subcateg = 'web'
for yr in range(2013,2018):
    yr_str = str(yr)
    date_range = yr_str + '-01-01 '+ yr_str+ '-12-31'
    pytrends.build_payload(kw_list, cat=0, timeframe=date_range, geo='SA', gprop=subcateg)
    ax = pytrends.interest_over_time().plot(figsize=(10,10))
    plt.ylabel('mean search rate',size = 20)
    plt.xlabel('month',size = 20)
    #ax.set_xticklabels([calendar.month_abbr[int(x.get_text())] for x in ax.get_xticklabels()])
    ax.grid(True)
    plt.xticks(size = 15)
    plt.yticks(size = 15)
    plt.legend('')
    plt.title('search statistics for '+ kw_list[0] + ' for year: ' + yr_str + " searched in " + subcateg,size = 20)
 
```

```python
def plot_keywrd_trend(kw_list,subcateg):

    df_data = pd.DataFrame(index=[i for i in range(1,13)])

    for yr in range(2010,2018):
        yr_str = str(yr)
        date_range = yr_str + '-01-01 '+ yr_str+ '-12-31'
        pytrends.build_payload(kw_list, cat=0, timeframe=date_range, geo='US', gprop=subcateg)
        df = pytrends.interest_over_time()
        df['month'] = pd.DatetimeIndex(df.index).month
        df_mnth=df.groupby('month').mean()
        df_mnth[kw_list[0]] = df_mnth[kw_list[0]]/df_mnth[kw_list[0]].max()
        df_data[yr_str] = df_mnth[kw_list[0]]

    df_data['mean'] =df_data.mean(axis = 1)
    df_data['std'] = df_data.std(axis = 1)
    ax = df_data.plot(y='mean',yerr='std',figsize=(12,12), marker ='o',markersize = 15)
    plt.ylabel('mean search rate',size = 20)
    plt.xlabel('month',size = 20)
    #ax.set_xticklabels([calendar.month_abbr[int(x.get_text())] for x in ax.get_xticklabels()])
    ax.grid(True)
    plt.xticks(size = 15)
    plt.yticks(size = 15)
    plt.legend('')
    plt.title('search statistics for '+ kw_list[0] + ' for sub_category: ' + subcateg,size = 20)
    #ax.set_xticklabels([calendar.month_abbr[int(x.get_text().strip())] for x in ax.get_xticklabels()])
```

```python
kwlist=['gun']
plot_keywrd_trend(kwlist,'')
```

```python
kwlist=['gun']
plot_keywrd_trend(kwlist,'froogle')
```

```python
kwlist=['gun']
plot_keywrd_trend(kwlist,'youtube')
```
